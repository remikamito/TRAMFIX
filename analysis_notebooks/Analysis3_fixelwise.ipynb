## **Set up and organise data**

### Step 1: Set up input and output directories

```python
import pandas as pd
import os

## Input dir containing data
input_dir_1 = '/Users/remika/Library/CloudStorage/OneDrive-TheUniversityofMelbourne/RESEARCH_PROJECTS/DECRA/TravellingHeads/Data/Analysis3_fixelwise/ICC'
input_dir_2 = '/Users/remika/Library/CloudStorage/OneDrive-TheUniversityofMelbourne/RESEARCH_PROJECTS/DECRA/TravellingHeads/Data/Analysis3_fixelwise/ICC_poptemp2'

## Output dir for plots
output_dir = '/Users/remika/Library/CloudStorage/OneDrive-TheUniversityofMelbourne/RESEARCH_PROJECTS/DECRA/TravellingHeads/Data/plots'
```

### Step 2: Import mrdump text files

```python
## POP TEMP 1 (site-specific processing) - FBA metrics

# FD ICC
icc_fd_df1 = os.path.join(input_dir_1, 'ICC_fd.txt')
icc_fd1 = pd.read_csv(icc_fd_df1, header=None, names=['ICC_FD'])

# FC ICC
icc_fc_df1 = os.path.join(input_dir_1, 'ICC_fc.txt')
icc_fc1 = pd.read_csv(icc_fc_df1, header=None, names=['ICC_FC'])

# FDC ICC
icc_fdc_df1 = os.path.join(input_dir_1, 'ICC_fdc.txt')
icc_fdc1 = pd.read_csv(icc_fdc_df1, header=None, names=['ICC_FDC'])

# FD CVws
cvws_fd_df1 = os.path.join(input_dir_1, 'CVws_fd.txt')
cvws_fd1 = pd.read_csv(cvws_fd_df1, header=None, names=['CVws_FD'])

# FC CVws
cvws_fc_df1 = os.path.join(input_dir_1, 'CVws_fc.txt')
cvws_fc1 = pd.read_csv(cvws_fc_df1, header=None, names=['CVws_FC'])

# FDC CVws
cvws_fdc_df1 = os.path.join(input_dir_1, 'CVws_fdc.txt')
cvws_fdc1 = pd.read_csv(cvws_fdc_df1, header=None, names=['CVws_FDC'])
```

```python
## POP TEMP 1 (site-specific processing) - DTI metrics

# FA ICC
icc_fa_df1 = os.path.join(input_dir_1, 'ICC_fa.txt')
icc_fa1 = pd.read_csv(icc_fa_df1, header=None, names=['ICC_FA'])

# FA skeleton ICC
icc_fa_skel_df1 = os.path.join(input_dir_1, 'ICC_fa_skel.txt')
icc_fa_skel_1 = pd.read_csv(icc_fa_skel_df1, header=None, names=['ICC_FA_skel'])

# ADC ICC
icc_adc_df1 = os.path.join(input_dir_1, 'ICC_adc.txt')
icc_adc1 = pd.read_csv(icc_adc_df1, header=None, names=['ICC_ADC'])

# ADC skeleton ICC
icc_adc_skel_df1 = os.path.join(input_dir_1, 'ICC_adc_skel.txt')
icc_adc_skel_1 = pd.read_csv(icc_adc_skel_df1, header=None, names=['ICC_ADC_skel'])

# FA CVws
cvws_fa_df1 = os.path.join(input_dir_1, 'CVws_fa.txt')
cvws_fa1 = pd.read_csv(cvws_fa_df1, header=None, names=['CVws_FA'])

# FA skeleton CVws
cvws_fa_skel_df1 = os.path.join(input_dir_1, 'CVws_fa_skel.txt')
cvws_fa_skel_1 = pd.read_csv(cvws_fa_skel_df1, header=None, names=['CVws_FA_skel'])

# ADC CVws
cvws_adc_df1 = os.path.join(input_dir_1, 'CVws_adc.txt')
cvws_adc1 = pd.read_csv(cvws_adc_df1, header=None, names=['CVws_ADC'])

# ADC skeleton CVws
cvws_adc_skel_df1 = os.path.join(input_dir_1, 'CVws_adc_skel.txt')
cvws_adc_skel_1 = pd.read_csv(cvws_adc_skel_df1, header=None, names=['CVws_ADC_skel'])
```

```python
## Pop temp 1 (ICC)
icc_fba_1 = pd.concat([icc_fd1, icc_fc1, icc_fdc1], axis=1)
icc_dti_1 = pd.concat([icc_fa1, icc_fa_skel_1, icc_adc1, icc_adc_skel_1], axis=1)
```

```python
## Pop temp 1 (CVws)
cvws_fba_1 = pd.concat([cvws_fd1, cvws_fc1, cvws_fdc1], axis=1)
cvws_dti_1 = pd.concat([cvws_fa1, cvws_fa_skel_1, cvws_adc1, cvws_adc_skel_1], axis=1)
```

### Step 3: Density plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# density plot
plt.figure(figsize=(6, 4))

for column in icc_fba_1.columns:
    sns.kdeplot(icc_fba_1[column], label=column, fill=True, alpha=0.5)  

for column in icc_dti_1.columns:
    sns.kdeplot(icc_dti_1[column], label=column, fill=True, alpha=0.5)

plt.legend()
plt.xlabel("ICC")
plt.ylabel("Density")
plt.tight_layout()

#plt.savefig(os.path.join(output_dir, "icc_density_plot.svg"), format="svg", bbox_inches="tight")

plt.show()
```
